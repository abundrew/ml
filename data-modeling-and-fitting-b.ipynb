# Data modeling and fitting

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

### Function  $ y = a\exp(\frac{-(x - \mu)^2}{2\sigma^2}) $
Create a function to model

```python
def func(x, a, b, c):
    return a * np.exp(-(x-b)**2/(2*c**2))
```

### Data
Generating clean data

```python
x = np.linspace(0, 10, 100)
y = func(x, 1, 5, 2)
```

### Noise
Adding noise to the data

```python
yn = y + 0.2 * np.random.normal(size=len(x))
```

### Data visualization
Plot data.

```python
plt.plot(x, yn, 'ro', label='data', ms=2.0)
plt.plot(x, func(x, 1, 5, 2), lw=2.0)
plt.show()
```

### Fit
Executing curve_fit on noisy data

```python
popt, pcov = curve_fit(func, x, yn)
```

### Best fit
Popt returns the best fit values for parameters of the given model (func).

```python
print(popt)
```

```python
plt.plot(x, yn, 'ro', label='Data', ms=2.0)
plt.plot(x, func(x, 1, 5, 2), label='Function', lw=2.0)
plt.plot(x, func(x, popt[0], popt[1], popt[2]), label='Best fit', lw=2.0)
plt.legend(loc='upper left', shadow=True)
plt.show()
```
